## 3. Parte 1 — Processamento básico de imagens e vídeos

### 3.A) Leitura de imagem de um arquivo

O programa `L__1_img.py` abre uma imagem que está salva no computador, mostra ela
em uma janela e ainda permite salvá-la em outro arquivo. Quando rodamos a versão
original, percebemos uma coisa curiosa: mesmo a foto sendo colorida, a janela
mostrava a imagem em **preto e branco**.

**Pergunta do roteiro:** *por que a janela aberta não mostra a imagem colorida?*

Isso acontece por causa do número `0` que estava sendo passado
na função `cv.imread`, na linha `img = cv.imread('messi5.jpg', 0)`. Esse segundo
argumento diz ao OpenCV de que jeito ler a imagem, e o valor `0` significa
"leia em tons de cinza". Ou seja, mesmo o arquivo sendo colorido, o programa o
convertia para preto e branco na hora de abrir, e era essa versão que aparecia
na tela.

Para corrigir e ver a imagem colorida, bastou **tirar esse `0`**: sem ele, o
OpenCV usa o modo padrão, que é o de cores. O código abaixo é a versão que
usamos:

```python
# Tiramos o "0" da leitura: agora a imagem é aberta em cores
img = cv.imread('messi5.jpg')
cv.imshow('image', img)
k = cv.waitKey(0)
if k == 27:          # tecla ESC: fecha a janela
    cv.destroyAllWindows()
elif k == ord('s'):  # tecla 's': salva e fecha
    cv.imwrite('messigray.png', img)
    cv.destroyAllWindows()
```

**O que mudamos:** trocamos `cv.imread('messi5.jpg', 0)` por
`cv.imread('messi5.jpg')`. Foi só essa pequena mudança que fez a imagem voltar
a aparecer colorida.

### 3.B) Leitura de um vídeo de arquivo

O programa `L__2_video.py` abre um vídeo já gravado (o `big_buck_bunny.mp4`) e vai
mostrando os quadros, um atrás do outro, numa janela. O roteiro pediu para
modificarmos o programa de modo que o vídeo passasse **mais rápido** e depois
**mais devagar**.

**Pergunta do roteiro:** *qual a explicação da alteração de velocidade?*

**Nossa resposta:** dentro do programa existe a linha `time.sleep(1/25.0)`, que
faz o programa **dar uma pausa antes de mostrar o próximo quadro**. Como o vídeo
é só uma sequência de imagens, é justamente o tamanho dessa pausa que decide a
velocidade que enxergamos:

- se **diminuirmos** a pausa, os quadros aparecem mais juntos e o vídeo fica
  **mais rápido**;
- se **aumentarmos** a pausa, os quadros demoram mais para trocar e o vídeo fica
  **mais lento**.

No código original a pausa era `1/25.0` (mais ou menos 25 quadros por segundo).
Para deixar mais rápido, usamos uma pausa menor, `1/50.0`:

```python
# Versão MAIS RÁPIDA: pausa menor (1/50) -> o vídeo corre mais rápido
cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv.imshow('frame', frame)
        time.sleep(1/50.0)   # pausa menor -> mais rápido
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv.destroyAllWindows()
```

E para deixar **mais lento**, usamos uma pausa maior, `1/10.0`:

```python
# Versão MAIS LENTA: pausa maior (1/10) -> o vídeo corre mais devagar
cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv.imshow('frame', frame)
        time.sleep(1/10.0)   # pausa maior -> mais lento
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv.destroyAllWindows()
```

**O que mudamos:** só o número dentro do `time.sleep`. Quanto menor esse
valor, mais rápido o vídeo passa; quanto maior, mais devagar. Resumindo, é como
ajustar o tempo de espera entre uma "foto" e outra.

### 3.C) Leitura de imagens da câmera

O programa `L__3_webcam.py` liga a webcam e fica mostrando ao vivo o que a câmera
está captando. O roteiro pediu para modificarmos o programa para que, ao
apertarmos a tecla **`x`**, ele **tirasse uma foto** (salvasse o quadro daquele
momento em um arquivo).

**Como fizemos:** o programa já lê continuamente qual tecla está sendo apertada.
Aproveitamos isso e dissemos ao programa que, quando a tecla apertada fosse o
`x`, ele deveria salvar o quadro atual em um arquivo chamado `foto_webcam.png` (usando
a função `cv.imwrite`), mostrar uma mensagem confirmando e encerrar. A tecla `q`
continuou servindo para fechar sem tirar foto.

```python
# Agora a webcam tira uma foto quando apertamos a tecla 'x'
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

while True:
    # captura um quadro
    ret, frame = cap.read()
    # se a leitura deu certo, ret é True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # mostra o quadro na janela
    cv.imshow('frame', frame)

    key = cv.waitKey(1)          # lê a tecla apertada
    if key == ord('x'):          # 'x': salva a foto e sai
        cv.imwrite('foto_webcam.png', frame)
        print("Foto salva como 'foto_webcam.png'!")
        break
    elif key == ord('q'):        # 'q': sai sem salvar
        break

cap.release()
cv.destroyAllWindows()
```

**O que mudamos:** acrescentamos o trecho que reconhece a tecla `x` e usa
o `cv.imwrite('foto1.png', frame)` para salvar a foto. Guardamos a tecla apertada
em uma variável (`key`) para depois comparar com `x` e com `q`, o que deixa o
programa mais confiável na hora de decidir o que fazer.

### 3.D) Gravação de um vídeo da câmera

O programa `L__4_webcap.py` é parecido com o anterior, mas em vez de só mostrar a
câmera ele **grava tudo em um arquivo de vídeo** (`saida.avi`), até apertarmos a
tecla `q`. Quando rodamos a versão original, notamos dois problemas: o vídeo
ficava **de cabeça para baixo** e a reprodução ficava meio "travada".

O roteiro pediu para deixarmos as imagens gravadas "normais" e com uma velocidade
adequada. Os dois ajustes que fizemos foram:

1. **Arrumar a orientação da imagem.** O programa usava `cv.flip(frame, 0)`, e
   esse `0` espelha a imagem na vertical, ou seja, deixa tudo de cabeça para
   baixo. Trocando o `0` por `1`, o espelhamento passa a ser na horizontal, que
   é o mesmo efeito de quando nos olhamos no espelho — a imagem fica na posição
   correta (em pé) e com aparência natural de webcam. (Se quiséssemos a imagem
   exatamente como a câmera vê, sem nenhum espelho, bastaria apagar essa linha.)

2. **Ajustar a velocidade do vídeo salvo.** O programa gravava com `fps = 10.0`,
   uma taxa baixa que deixa o vídeo lento. Aumentamos para `fps = 30.0`, que é
   uma taxa comum de webcams e deixa a reprodução bem mais fluida.

Ficou assim:

```python
# Vídeo gravado com a orientação correta (flip 1) e velocidade boa (30 fps)
cap = cv.VideoCapture(0)

# pega a largura e a altura atuais do quadro
width = cap.get(cv.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT)
# taxa de quadros do vídeo
fps = 30.0   # antes era 10.0 -> agora a reprodução fica mais fluida

# define o "formato" do vídeo e cria o gravador
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('saida.avi', fourcc, fps, (int(width), int(height)))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    frame = cv.flip(frame, 1)   # antes era 0 (de cabeça p/ baixo); agora fica em pé
    out.write(frame)            # grava o quadro no arquivo
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
out.release()
cv.destroyAllWindows()
```

**O que mudamos:** trocamos `cv.flip(frame, 0)` por `cv.flip(frame, 1)`
(para a imagem ficar em pé) e o `fps` de `10.0` para `30.0` (para o vídeo ficar
com velocidade normal). O resultado é gravado no arquivo `saida.avi`.

### 3.E) Onde fazer alterações na imagem?

Para fechar a Parte 1, o roteiro fez uma pergunta mais geral:

**Pergunta:** *se for preciso alterar a imagem, ou seja, fazer alguma operação de
processamento nela, em que ponto dos quatro programas isso deve ser feito?*

**Nossa resposta:** a alteração deve ser feita **depois de ler a imagem e antes
de mostrá-la ou gravá-la** — bem no meio do caminho "entrada → processamento →
saída". Nos programas de câmera e de vídeo, isso quer dizer logo depois da linha
que lê o quadro (`ret, frame = cap.read()`) e antes das linhas que mostram
(`cv.imshow`) ou que gravam (`out.write`), dentro do laço de repetição. Foi
exatamente nesse lugar que colocamos o `cv.flip` no item D. No programa de
imagem, o raciocínio é o mesmo: a alteração entra depois do `cv.imread` e antes
do `cv.imshow`.

Em palavras simples: primeiro pegamos a imagem, depois fazemos o que quisermos
com ela, e só então mostramos ou salvamos o resultado.

### 4.c) Gravação dos vídeos

Por fim, gravamos **dois vídeos com pessoas** e **dois vídeos com um objeto**.
Em cada par, um vídeo tem **movimento lento** e o outro tem **movimento rápido**,
e os integrantes que aparecem em cada vídeo são diferentes.

Todos os vídeos foram gravados com o mesmo programa corrigido no item 3.D (com a
imagem em pé e 30 fps); a única coisa que mudamos de um para o outro foi o **nome
do arquivo** de saída. As células abaixo mostram cada gravação.

**Vídeo 1 — pessoa, movimento lento:**

```python
# Vídeo 1 - PESSOA com movimento LENTO
cap = cv.VideoCapture(0)

width = cap.get(cv.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT)
fps = 30.0

fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('duas_pessoas_movimento_lento.avi', fourcc, fps, (int(width), int(height)))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    frame = cv.flip(frame, 1)
    out.write(frame)
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
out.release()
cv.destroyAllWindows()
```

**Vídeo 2 — pessoa, movimento rápido:**

```python
# Vídeo 2 - PESSOA com movimento RÁPIDO
cap = cv.VideoCapture(0)

width = cap.get(cv.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT)
fps = 30.0

fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('duas_pessoas_movimento_rapido.avi', fourcc, fps, (int(width), int(height)))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    frame = cv.flip(frame, 1)
    out.write(frame)
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
out.release()
cv.destroyAllWindows()
```
